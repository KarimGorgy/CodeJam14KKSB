```python
def recommend_cars(
    df,
    car_type=None,          # e.g., 'SUV', 'Sedan'
    price_max=None,         # e.g., 30000
    year_min=None,          # e.g., 2020
    mileage_max=None,       # e.g., 50000
    color=None,             # e.g., 'Red'
    fuel_type=None          # e.g., 'Gasoline', 'Electric'
):
    filtered_df = df.copy()
    
    # Filter by car type (Body)
    if car_type:
        filtered_df = filtered_df[filtered_df['Body'].str.contains(car_type, case=False, na=False)]
    
    # Filter by maximum price
    if price_max:
        filtered_df = filtered_df[filtered_df['SellingPrice'] <= price_max]
    
    # Filter by minimum year
    if year_min:
        filtered_df = filtered_df[filtered_df['Year'] >= year_min]
    
    # Filter by maximum mileage
    if mileage_max:
        filtered_df = filtered_df[filtered_df['Miles'] <= mileage_max]
    
    # Filter by exterior color
    if color:
        filtered_df = filtered_df[filtered_df['ExteriorColor'].str.contains(color, case=False, na=False)]
    
    # Filter by fuel type
    if fuel_type:
        filtered_df = filtered_df[filtered_df['Fuel_Type'].str.contains(fuel_type, case=False, na=False)]
    
    # Sort by price, year, and mileage for better ranking
    filtered_df = filtered_df.sort_values(by=['SellingPrice', 'Year', 'Miles'])
    
    # Return top 5 matches
    return filtered_df.head(5)
```

```python
import spacy
import re

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def parse_user_input(user_input):
    doc = nlp(user_input.lower())
    
    # Default preferences
    preferences = {
        "car_type": None,
        "price_max": None,
        "mileage_max": None,
        "color": None,
        "year_min": None
    }
    
    # Extract car type
    car_types = ["suv", "sedan", "truck", "convertible", "coupe"]
    for token in doc:
        if token.text in car_types:
            preferences["car_type"] = token.text
    
    # Extract price (e.g., "under $30K")
    price_match = re.search(r"under \$?(\d+)[kK]?", user_input)
    if price_match:
        preferences["price_max"] = int(price_match.group(1)) * 1000
    
    # Extract year (e.g., "2020 or newer")
    year_match = re.search(r"(\d{4}) or newer", user_input)
    if year_match:
        preferences["year_min"] = int(year_match.group(1))
    
    # Extract color
    colors = ["red", "blue", "black", "white", "silver"]
    for token in doc:
        if token.text in colors:
            preferences["color"] = token.text
    
    return preferences
```
