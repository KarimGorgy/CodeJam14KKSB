```python
from spacy.training import Example

training_data = []
for text, annotations in examples:
    doc = nlp.make_doc(text)
    example = Example.from_dict(doc, annotations)
    training_data.append(example)
```

```python
import random
from spacy.util import minibatch, compounding
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
    # Training code here
# Begin training
with nlp.disable_pipes(*other_pipes):
    optimizer = nlp.create_optimizer()
    for epoch in range(50):  # Adjust the number of epochs as needed
        random.shuffle(training_data)
        losses = {}
        batches = minibatch(training_data, size=compounding(4.0, 32.0, 1.5))
        for batch in batches:
            nlp.update(
                batch,
                sgd=optimizer,
                drop=0.35,
                losses=losses
            )
        print(f"Epoch {epoch + 1}, Loss: {losses['ner']:.4f}")
```

```python
import pandas as pd
import random

# Assuming df is your dataframe
df = pd.read_csv("./../vehicles.csv")

# Extract unique values
makes = df['Make'].dropna().unique().tolist()
models = df['Model'].dropna().unique().tolist()
years = df['Year'].dropna().unique().astype(str).tolist()
prices = df['Selling_Price'].dropna().unique().tolist()
colors = df['Color'].dropna().unique().tolist()
body_types = df['Body_Type'].dropna().unique().tolist()
```

```python

# Define templates
templates = [
    "I am interested in a {year} {make} {model}.",
    "Do you have any {color} {body_type}s available?",
    "Looking for a {make} priced under ${price}.",
    "Can you find me a {make} {model} with less than {mileage} miles?",
    "I need a {fuel_type} {body_type} with {options}."
]

synthetic_examples = []

for _ in range(1000):
    template = random.choice(templates)
    sentence = template.format(
        year=random.choice(years),
        make=random.choice(makes),
        model=random.choice(models),
        price=str(random.randint(10000, 50000)),
        color=random.choice(colors),
        body_type=random.choice(body_types),
        mileage=str(random.randint(10000, 100000)),
        fuel_type=random.choice(df['Fuel_Type'].dropna().unique().tolist()),
        options=random.choice(df['Options'].dropna().unique().tolist())
    )
    # Annotation logic here...
    # You can write a function to find the start and end indices of each entity in the sentence
    entities = []
    for label, value in [("YEAR", years), ("MAKE", makes), ("MODEL", models), ("PRICE", prices),
                         ("COLOR", colors), ("BODY", body_types)]:
        for val in value:
            if val in sentence:
                start = sentence.index(val)
                end = start + len(val)
                entities.append((start, end, label))
    synthetic_examples.append((sentence, {"entities": entities}))
```
